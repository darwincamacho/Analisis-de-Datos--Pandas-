## Principales opciones de configuracion

- Especificar el nombre del archivo: pd.read_csv('nombre.csv')
- Especificar el delimitador: pd.read_csv('nombre.csv',sep(;))
- Especificar el encabezado: pd.read_csv('nombre.csv',header=None)
- Especificar columnas a trabajar: pd.read_csv('nombre.csv',names=['col1','col2'])
- Especificar la codificacion: pd.read_csv('nombre.csv',encoding='utf-8')
- Especificar el n√∫mero de filas: pd.read_csv('nombre.csv',nrows=10)
- Especificar si tienes valores nulos o faltantes (NaN): pd.read_csv('nombre.csv',na_Values=['N/A','-'])
- Especificar formato de columna: pd.read_csv('nombre.csv',dtype={'col1':str})

```python
import pandas as pd

"""Leer el archivo csv de Lista_empleados"""

df=pd.read_csv(
    '../Bases_datos/Lista_empleados.csv',sep=';'
    )
df
```

```python
"""Reemplazar Sin Datos en la columna Nombre"""

df[['Nombre']]=df[['Nombre']].fillna('Sin Datos')
```

```python
"""la columna fecha la tenemos que especificar con formato de fecha"""

df['Fecha']=pd.to_datetime(df['Fecha'],format="%Y-%m-%d",errors='coerce')
df
```

```python
"""Formatear edad y salario a numeros"""

df['Edad']=pd.to_numeric(df['Edad'],errors='coerce')
df['Salario']=pd.to_numeric(df['Salario'],errors='coerce')
```

```python
"""rellenar con 0 las columnas numericas ya que al formatear la columna fecha ya se puson en NaT"""

df[['Edad','Salario']]=df[['Edad','Salario']].fillna(0)
df
```

```python
"""descargar el archvo csv modificado"""

df.to_csv('archivo_limpio.csv')
```
