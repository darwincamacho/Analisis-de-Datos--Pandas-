| Símbolo             | Uso principal en Pandas                              | Ejemplos correctos                                                                                                                        | Qué hace                                                |
| ------------------- | ---------------------------------------------------- | ----------------------------------------------------------------------------------------------------------------------------------------- | ------------------------------------------------------- |
| **`()`** paréntesis | Ejecutar un **método** de un DataFrame o Serie       | `df.head(5)`<br>`df.describe()`<br>`df.groupby("col").sum()`                                                                              | Llama a la función/método y devuelve un resultado       |
| **`[]`** corchetes  | **Seleccionar datos** (columnas, filas, condiciones) | `df["col"]` → una columna (Serie)<br>`df[["col1","col2"]]` → varias columnas (DataFrame)<br>`df[df["col"] > 10]` → filtrado por condición | Extrae datos por etiqueta o posición                    |
| **`{}`** llaves     | Diccionarios para mapeo o creación de DataFrames     | `pd.DataFrame({"A":[1,2], "B":[3,4]})`<br>`df.replace({"A": {"old": "new"}})`                                                             | Estructuras clave-valor usadas en reemplazos o creación |


```python
"""METODO INDEX: indicar a python que la comuna indice es la columna indice"""

#df_vic=df_vic.set_index('index') #dice index porque el nombre de la columna realmente se llama indix pero podria ser cualquier nombre
```

```python
"""METODO HEAD: filtrar por numero de columnas, Ej: quiero ver las 10 primeras"""

df_vic.head(10)
```

```python
"""METODO INFO: muestra los nombres de las columnas"""

df_vic.info()
```

```python
"""INDEXADOR LOC () - atributo especial accessor: localizar y filtra los indices indicados (va entre corchetes porque no es una funcion,
es un idexador) (tambien es necesario escribir la funcion set_index porque python por default comienza por 0 y por lo general las listas
comienzan en 1)"""

df_vic.loc[2]
```

```python
"""INDEXADOR LOC: encontrar varios indices a la vez (usa doble corchete ya que le vas a pasar una lsita)"""

df_vic.loc[[2,3,4,5]]
```

```python
"""FUNCION UNIQUE: filtra solo una columna (va en corchetes porque los corchetes tienen la funcion de seleccionar datos)"""

df_vic['Urban_area_type'].unique()
```

```python
"""FUNCION LOC + datos de columna (filtrar los datos dentro de una columna) esto es como indice y coincidir en excel"""

df_vic.loc[df_vic['Urban_area_type']=='Main urban area']

#Obtendrás todas las filas donde Urban_area_type es 'Main urban area'.
#es como aplicar filtros sobre una columna en excel
```

```python
"""ATRIBUTO COLUMNS: devuelve el nombre de las columnas"""

df_vic.columns
```

```python
"""FILTRO POR COLUMNAS (solo las 10 primeras)"""

df_vic[['Area_unit_2013_code','Area_unit_2013_label']].head()
```

```python
"""METODO ISNA (NaN - que faltan registros o vacios)"""

df_vic.isna()
```

```python
"""METODO ISNA + FILTRO DE COLUMNA"""

df_vic[df_vic['Area_unit_2013_code'].isna()]

#el resultado indica que no hay nunguna columa en nan o null 
```

```python
"""METODO ISIN: filtra los valores de una colmna
1.- nombre de la columna
2.- nombre de la nueva lista que quieres filtrar"""

newdf=['Auckland Central West', 'Hamilton Central']

df_vic[df_vic['Area_unit_2013_label'].isin(newdf)]

#se crea una lista nueva para no tener que escribir los nombres de nuevo
```

## signos para unir varias condiciones logicas al mismo tiempo

- & (and) Se cumplen ambas condiciones
- | (or) Se cumple al menos una condicion
- ~ (not) Negacion

```python
df_vic[(df_vic['Urban_area_type']=='Secondary urban area') & (df_vic['Victimisations_calendar_year_2015']<253)]
```

```python
df_vic[(df_vic['Urban_area_type']=='Secondary urban area') | (df_vic['Victimisations_calendar_year_2015']<253)]
```

```python
df_vic[(df_vic['Urban_area_type']=='Secondary urban area') ~ (df_vic['Victimisations_calendar_year_2015']<253)]

```

PORQUE SE USA ENCODING: porque sale el siguiente eror: Ese UnicodeDecodeError te está diciendo básicamente:

"Pandas intentó leer el CSV asumiendo que está en UTF-8, pero encontró caracteres que no cuadran con esa 
codificación".

Esto es muy común cuando trabajas con archivos CSV que vienen de Excel, gobiernos o sistemas antiguos, porque 
suelen estar en Latin-1 (ISO-8859-1) o Windows-1252 en lugar de UTF-8.

Si el archivo tiene acentos, tildes, ñ, u otros caracteres especiales guardados con otra codificación, Python 
se confunde y lanza UnicodeDecodeError.

Usar latin1 o ISO-8859-1 le dice a Pandas cómo interpretar esos bytes.

y el SEP= es para convertir ";" en comas ya que el formato csv generalmente esta separado por comas, y si 
en el archivo esta separado por ; (dependiendo de la version de excel) entonces hay que convertir
