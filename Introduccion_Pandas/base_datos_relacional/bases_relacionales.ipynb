## Manejo de base de datos realcionales



```python
"""Configurar la base de datos de postgresql a sqlalchemy"""

from sqlalchemy import create_engine
```

```python
"""configurar los parámetros de conección a postgresql"""

username='postgres'
password='7138'
host='localhost'
port='5432'
database='productos'

"""Crear la url de conección"""

engine=create_engine(f'postgresql+psycopg2://{username}:{password}@{host}:{port}/{database}')
```

```python
"""Ejecutar conusltas SQL"""

import pandas as pd

#primero se crea la variable query
query='SELECT * FROM productos'

#se crea el DataFrame (con=engine "Especifica la conexión a la base de datos creada previamente"); defina la columna id como pk
df=pd.read_sql_query(query,con=engine, index_col='id')

#imprimir
df
```

```python
"""filtrar los productos que son de categoria_id = 1"""

query2='SELECT * FROM productos WHERE categoria_id = 1'
df2=pd.read_sql_query(query2,con=engine, index_col='id')
df2
```

```python
"""seleccionar la tabla categorias"""

query3 = """ SELECT *
    FROM categorias
"""
df3 = pd.read_sql_query(query3,con=engine,index_col='categoria_id')
df3    
```

```python
query4="""
    SELECT *
    FROM productos p
    JOIN categorias c
    ON p.id = c.categoria_id
"""

df4=pd.read_sql_query(query4, con=engine)
df4
```

```python
"""Insertar datos desde un DataFrame de pandas a pgAdmin"""

df_nuevo=pd.DataFrame({
    'nombre':['algarrobina'],
    'estado':[True]
})

#Insertar el nuevo daataframe a la tabla categoria que esta en psadmin
df_nuevo.to_sql('categorias',con=engine,if_exists='append',index=False)
```

```python
#revisar si se agrego, llamar a categoria
query3="""
    SELECT *
    FROM categorias
"""
df3=pd.read_sql_query(query3,con=engine)
df3
```

```python
"""Modificar datos de las tablas con update y la libreria text de sqlalchimy"""

from sqlalchemy import text

with engine.connect() as conn:
    conn.execute(
        text("""
                UPDATE categorias
                SET nombre = 'Algarrobina'
                WHERE nombre = 'algarrobina'
        """)
    )
    conn.commit()
```
