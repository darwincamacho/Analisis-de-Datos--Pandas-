## Graficos con matplotlib

```python
import matplotlib.pyplot as plt
```

```python
"""Grafico de barras"""

categorias=['A','B','C','D']
valores=[4,7,1,8]

plt.figure(figsize=(7,4)) #lado y altura
plt.bar(categorias,valores,color='blue')
plt.show()
```

```python
#Al mismo grafico agregar titulo y etiquetas

plt.figure(figsize=(7,4)) #lado y altura
plt.bar(categorias,valores,color='blue')
#agreagar titulo y coordenadas de la etiqueta de los ejes
plt.title('GRAFICO DE BARRAS')
plt.xlabel('Categorias')
plt.ylabel('Valores')

plt.show()
```

```python
"""Grafico de lineas"""

x=[1,2,3,4,5]
y=[2,3,5,7,11]

plt.figure(figsize=(7,4)) #lado y alto
plt.plot(x,y,marker='o',markerfacecolor='black',color='orange',linestyle='-',linewidth=1) #linewidth se utuliza para aumentar el ancho de el objeto
plt.title('GRAFICO DE LINEAS')
plt.xlabel('X')
plt.ylabel('Y')
plt.grid()#agregar cuadricula
plt.show()
```

```python
"""Histograma"""

#generaremos datos aleatorios por lo que necesitamos importa la libreria de numoy
import numpy as np
#generacion de datos aleatorios
data=np.random.randn(1000) #1000 numeros aleatorios

#Histograma
plt.figure(figsize=(7,4))
plt.hist(data,bins=30,color='green',alpha=0.7) #probar con bins 10 y 100 (un ao) y alpha es la trasparencia de las barras
plt.title('HISTOGRAMA')
plt.xlabel('Valores')
plt.ylabel('Frecuencia')
plt.show() #como se puede ver , los datos tienen una distribucion normal
```

 bins=30

Define en cuntos intervalos (cajones o barras) se dividirn los datos.

Es literalmente cmo se agrupa la distribucin.

Ejemplo:

Si tienes edades de personas de 0 a 100:

bins=10  te genera intervalos de 10 en 10 (010, 1020, , 90100).

bins=100  cada intervalo sera de 1 ao (01, 12, , 99100).

 Traduccin corporativa: Ms bins = ms detalle (microsegmentacin). Menos bins = visin macro (agregados grandes).
