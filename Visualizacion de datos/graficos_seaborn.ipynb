## GRAFICOS CON SEABORN


```python
import matplotlib.pyplot as plt
import seaborn as sns
import pandas as pd
import numpy as np
```

```python
"""Crear un DataFrame simple como origen de datos"""

data=pd.DataFrame({ #Diccionario= Clave:Valor
    'x':np.arange(1,11), #crea arrays (arange), valores del 1 al 11
    'y':np.random.randint(5,20,10), #10 valore aleatorios entre 5 y 20
    'category':np.random.choice(['A','B'],size=10)
})
data
```

```python
sns.lineplot(x='x',y='y',data=data,marker='o')
plt.title('GRAFICO DE LINEAS SIMPLE')
plt.xlabel('Eje x')
plt.ylabel('Eje y')
plt.show()
```

```python
"""crear un grafico de barras"""

sns.barplot(x='x',y='y',data=data,palette='pastel')
plt.title('GRAFICO DE BARRAS SIMPLE')
plt.xlabel('Eje X')
plt.ylabel('Eje Y')
plt.show()
```

```python
"""Grafico de dispersion"""

sns.scatterplot(x='x',y='y',data=data,hue='category',style='category',palette='muted',s=40) 
plt.title('GRAFICO DE DISPERSION')
plt.xlabel('Eje X')
plt.ylabel('Eje Y')
plt.show()
```

Explicacin de cada parmetro

x='x'
 Columna del DataFrame data que usas en el eje X.
Aqu graficas la columna llamada x.

y='y'
 Columna del DataFrame data que usas en el eje Y.
Aqu graficas la columna llamada y.

data=data
 Le dices a Seaborn de dnde sacar las columnas (x y y).
En este caso, tu DataFrame data.

hue='category'
 Colorea los puntos segn la columna category.
Ejemplo: los de categora A en azul, los de B en naranja.

style='category'
 Adems de color, cambia el marcador segn la categora.
Ejemplo: A = crculo, B = tringulo.

palette='muted'
 Define la paleta de colores. "muted" usa tonos suaves (profesionales, menos saturados).
Otras opciones: "deep", "pastel", "bright", "dark", "colorblind".

s=100
 Tamao de los puntos (scatter size).
El default suele ser 40, aqu lo aumentaste para que se vean ms grandes.

 Visualmente, qu obtienes?

Un grfico de dispersin con eje X = x y eje Y = y.

Cada punto representa una fila de tu DataFrame.

El color y el estilo de marcador dependen de la columna category.

Colores suaves por la paleta "muted".

Puntos de tamao mediano-grande (100).
