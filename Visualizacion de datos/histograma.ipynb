## HISTOGRAMA

Un histograma es un tipo de gráfico que utiliza barras para mostrar la distribución de frecuencia de un conjunto de datos numéricos continuos. Cada barra representa un rango o intervalo de valores, y su altura indica cuántos puntos de datos caen dentro de ese intervalo. Los histogramas ayudan a visualizar patrones, identificar valores atípicos y entender la forma general de los datos. 

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
```

```python
#Generar datos aleatorios con las califiaciones de 100 estudiantes
calificaciones=np.random.normal(loc=75,scale=10,size=100)

#Asegurar califiaciones que esten entre 0 - 100
calificaciones=np.clip(calificaciones,0,100)
calificaciones
```

```python
#crear el histograma
plt.figure(figsize=(10,5))
plt.hist(calificaciones,bins=10,color='skyblue',edgecolor='black',alpha=0.7) #hist de histograma
plt.title('DISTRIBUCION NORMAL DE CALIFICACIONES')
plt.xlabel('Calificaciones')
plt.ylabel('Número de calificaciones')
plt.grid(axis='y',alpha=0.75) #lineas horizontales, trasparencia del grafico 70% solido y 30% trasparencia
plt.show()
```

Estamos viendo la frecuencia de las calificaciones, podemos que el nuemro de calificaciones se repite con mas frecuencia en la nota 75 (rango 75)
