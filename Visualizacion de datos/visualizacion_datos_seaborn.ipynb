## VISUALIZACION DE DATOS CON SEABORN 2

```python
import seaborn as sns
import matplotlib.pyplot as plt
import pandas as pd
import numpy as np
```

```python
#configurar el estilo de seaborn

sns.set(style='whitegrid')
```

```python
#crear un DataFrame con datos aleatorios
np.random.seed(42) #fija los numeros aleatorios

data=pd.DataFrame({ #clave:valor
    'category':np.random.choice(['A','B','C'],size=100),
    'valor1':np.random.normal(10,5,100), #np.random.normal(loc, scale, size) 
    'valor2':np.random.normal(10,15,100),
    'valor3':np.random.normal(10,30,100)
})
data.head()

#üìå Sintaxis de np.random.normal(loc, scale, size)

#loc=10 ‚Üí la media de la distribuci√≥n normal (el centro).

#scale=5 ‚Üí la desviaci√≥n est√°ndar (qu√© tan dispersos est√°n los datos).

#size=100 ‚Üí cu√°ntos n√∫meros aleatorios quieres generar.
```

üìå ¬øQu√© hace np.random.seed(42)?

np.random es el m√≥dulo de numpy para generar n√∫meros aleatorios.

seed(42) establece una semilla fija en el generador de aleatorios.

üëâ Eso significa que cada vez que ejecutes tu c√≥digo, los "n√∫meros aleatorios" generados ser√°n los mismos.

ü§î ¬øPor qu√© se usa mucho el n√∫mero 42?

Es una especie de ‚Äúinside joke‚Äù en la comunidad cient√≠fica/tech.

Referencia al libro ‚ÄúThe Hitchhiker‚Äôs Guide to the Galaxy‚Äù, donde 42 es ‚Äúla respuesta a la vida, el universo y todo lo dem√°s‚Äù.

Se volvi√≥ el seed favorito para ejemplos en docs, notebooks, y papers

```python
sns.pairplot(data,hue='category',palette='Dark2',diag_kind='kde') #kde=curvas de densidad (Kernel Density Estimation)
#mostrar el grafico
plt.suptitle('VISUALIZACION DE PARES POR CATEGORIA',y=1.02)
plt.show()
```

```python
#calcular la corelacion
correlation=data[['valor1','valor2','valor3']].corr()
#crear una mascara para ocultar la mitad superior del mapa
mask=np.triu(np.ones_like(correlation,dtype=bool))
#crear el mapa de calor
plt.figure(figsize=(10,6))
sns.heatmap(correlation,annot=True,cmap='coolwarm',mask=mask,linewidths=0.5,linecolor='white',cbar_kws={'shrink':0.8})
plt.suptitle('MAPA DE CALOR DE CORRELACION CON MASCARA')
plt.show()
```
