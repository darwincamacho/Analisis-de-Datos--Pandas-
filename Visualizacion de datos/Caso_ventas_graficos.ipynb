## caso practico de ventas con graficos 

creacion de base de datos con mumeros, fechas y productos random

```python
import matplotlib.pyplot as plt
import numpy as np
import pandas as pd
```

¬øQu√© significa la np.random.seed(42)?

np.random es el m√≥dulo de NumPy que genera n√∫meros aleatorios.
.seed(42) configura la semilla (seed) del generador de n√∫meros pseudoaleatorios.
Eso asegura que cada vez que ejecutes tu c√≥digo, los n√∫meros generados sean los mismos.
En otras palabras: fija el ‚Äúpunto de partida‚Äù de la aleatoriedad.
(42) En realidad podr√≠as usar cualquier entero (ej: 123, 2025, etc.).
El 42 es una convenci√≥n geek ü§ì tomada de ‚ÄúLa Gu√≠a del Autoestopista Gal√°ctico‚Äù, donde ‚Äú42 es la respuesta al sentido de la vida, el universo y todo lo dem√°s‚Äù.

```python
#configuar la semilla para la reproductibilidad
np.random.seed(42)
#fechas para el mes de enero
fechas=pd.date_range(start='2024-01-01', end='2024-01-31')
fechas
```

```python
#generar datos de ventas aleatorias
ventas=np.random.randint(50.200,size=len(fechas))
ventas
```

```python
#generar productos
productos=np.random.choice(['PRODUCTO A','PRODUCTO B', 'PRODUCTO C'],size=len(fechas))
productos
```

```python
#Crear el DataFrame
df = pd.DataFrame(
    {
        'fecha':fechas,
        'venta':ventas,
        'producto':productos
    }
)
df
```

```python
#Sumar las ventas y agrupar por fechas
ventas_diarias=df.groupby('fecha')['venta'].sum().reset_index()
ventas_diarias
```

```python
#Crar el grafico para las ventas diarias (las fechas van con grafico de lineas por convencion)
plt.figure(figsize=(10,5))
plt.plot(ventas_diarias['fecha'],ventas_diarias['venta'],marker='o',color='blue',linestyle='-')
plt.show()
```

```python
"""Fine Tuning (Formatear el grafico, titulo, etiquetas y el leyout)"""
plt.figure(figsize=(10,5))
plt.plot(ventas_diarias['fecha'],ventas_diarias['venta'],marker='o',color='blue',linestyle='-')
# Recorremos cada punto y le ponemos el valor
for x, y in zip(df["fecha"], df["venta"]):
    plt.text(x, y+0.5, str(y), ha='center', va='bottom', fontsize=8) #este arreglo se agrego la final
plt.title('VENTAS DIARIAS, ENERO 2024')
plt.xlabel('Fecha')
plt.ylabel('Ventas')
plt.xticks(rotation=45)
plt.grid(True) #cuadricula interior
plt.tight_layout()#redimensiona para que todo este en su lugar
plt.show()
```

Explicaci√≥n:

x, y ‚Üí coordenadas de cada punto.

y+0.5 ‚Üí para que el n√∫mero no quede encima del punto, sino un poquito arriba.

ha='center' ‚Üí alineaci√≥n horizontal al centro.

va='bottom' ‚Üí alineaci√≥n vertical desde abajo.

fontsize=8 ‚Üí tama√±o peque√±o para que no se vea saturado.

```python
"""Ventas totales por producto"""

#agrupar por prodcuto y luego agrupar las ventas
venta_producto=df.groupby('producto')['venta'].sum().reset_index()
venta_producto
```

```python
#Grafico de barras
plt.figure(figsize=(8,5))
plt.bar(venta_producto['producto'],venta_producto['venta'],color='green',alpha=0.7) #En Matplotlib, el par√°metro alpha controla la transparencia (u opacidad) de los elementos gr√°ficos.
plt.title('VENTAS TOTALES POR PRODUCTO')
plt.xlabel('Producto')
plt.ylabel('Venta')
plt.tight_layout()
plt.show()
```
