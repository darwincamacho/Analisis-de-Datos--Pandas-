| Símbolo             | Uso principal en Pandas                              | Ejemplos correctos                                                                                                                        | Qué hace                                                |
| ------------------- | ---------------------------------------------------- | ----------------------------------------------------------------------------------------------------------------------------------------- | ------------------------------------------------------- |
| **`()`** paréntesis | Ejecutar un **método** de un DataFrame o Serie       | `df.head(5)`<br>`df.describe()`<br>`df.groupby("col").sum()`                                                                              | Llama a la función/método y devuelve un resultado       |
| **`[]`** corchetes  | **Seleccionar datos** (columnas, filas, condiciones) | `df["col"]` → una columna (Serie)<br>`df[["col1","col2"]]` → varias columnas (DataFrame)<br>`df[df["col"] > 10]` → filtrado por condición | Extrae datos por etiqueta o posición                    |
| **`{}`** llaves     | Diccionarios para mapeo o creación de DataFrames     | `pd.DataFrame({"A":[1,2], "B":[3,4]})`<br>`df.replace({"A": {"old": "new"}})`                                                             | Estructuras clave-valor usadas en reemplazos o creación |

PORQUE SE USA ENCODING: porque sale el siguiente error: Ese UnicodeDecodeError te está diciendo básicamente:

"Pandas intentó leer el CSV asumiendo que está en UTF-8, pero encontró caracteres que no cuadran con esa 
codificación".

Esto es muy común cuando trabajas con archivos CSV que vienen de Excel, gobiernos o sistemas antiguos, porque 
suelen estar en Latin-1 (ISO-8859-1) o Windows-1252 en lugar de UTF-8.

Si el archivo tiene acentos, tildes, ñ, u otros caracteres especiales guardados con otra codificación, Python 
se confunde y lanza UnicodeDecodeError.

Usar latin1 o ISO-8859-1 le dice a Pandas cómo interpretar esos bytes.

y el SEP= es para convertir ";" en comas ya que el formato csv generalmente esta separado por comas, y si 
en el archivo esta separado por ; (dependiendo de la version de excel) entonces hay que convertir

```python
"""FILTRAR POR DEPARTAMENTO, LOS 5 DEPARTAMENTOS CON MAS FALLECIDOS"""

#df_at['DEPARTAMENTO'].value_counts().head()

#el resultado dice: hubo n registros por departamento, pero no da el numero de fallecidos, tenemos que usar "groupby"
```

```python
"""METODO GROUPBY: método le permite agrupar sus datos y ejecutar funciones en estos grupos."""

top5_departamentos_con_mas_fallecidos=(
    df_at.groupby('DEPARTAMENTO', as_index=False)['FALLECIDOS']
    .sum()
    .sort_values('FALLECIDOS', ascending=False)
    .head()
)
print(top5_departamentos_con_mas_fallecidos)
```

```python
"""El error es clasico, nos dice que la columna de fallecidos no es de tipo numerica
tenemos que convertir la columna a formato numeros primero antes de hacer los filtros correspondientes"""

import pandas as pd

df_at=pd.read_csv(
    '../Bases_datos/Accidentes_tránsito_carretera-Sutran.csv',encoding='latin1',sep=';',index_col='INDICE'
)
df_at
```

```python
"""filtrar los 5 departamentos con mas fallecidos"""

top5_dep_fallecidos=(
    df_at.groupby('DEPARTAMENTO', as_index=False)['FALLECIDOS']
    .sum()
    .sort_values('FALLECIDOS', ascending=False)
    .head(5)
)
top5_dep_fallecidos
```

```python
"""para que no aparesca ese error hay que convertir la columna fallecidos a nueros,
paorque tal parace que los esta tomando como strings y esta sumando todos los caracteres"""

df_at['FALLECIDOS']=pd.to_numeric(df_at['FALLECIDOS'],errors='coerce')
```

```python
"""Ahora volvemos a correr el script para ver el top 5 de fallecidos"""

top5_dep_fallecidos=(
    df_at.groupby('DEPARTAMENTO', as_index=False)['FALLECIDOS']
    .sum()
    .sort_values('FALLECIDOS',ascending=False)
    .head(5)
)
top5_dep_fallecidos
```

```python
"""otro metodo mas compacto con 'nlargest(n,'nombre_columna)"""

top5_dep_fallecidos=(
    #agrupamos por departamento y sumamos por fallecidos
    df_at.groupby('DEPARTAMENTO', as_index=False)['FALLECIDOS']
    .sum()
    # Top 5 con nlargest (más directo y eficiente)
    .nlargest(5,'FALLECIDOS')
)
top5_dep_fallecidos
```

```python
"""localiza los 5 fallecimientos con mas de una persona en lima"""

df_at.loc[(df_at['DEPARTAMENTO']=='LIMA') & (df_at['FALLECIDOS']>1)].head(5)
```

```python
"""para descargar el archivo como un nuevo 'csv' """

#se crea una variable a partir del resultado
df_result=df_at.loc[(df_at['DEPARTAMENTO']=='LIMA') & (df_at['FALLECIDOS']>1)].head(5)
df_result.to_csv('primer_csv')
```
